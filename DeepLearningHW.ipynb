### **Preprocessing**

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/mel518/DeepLearning/main/charity_data.csv"
charity_data_df = pd.read_csv(url)
charity_data_df.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
```

**Target** = IS_SUCCESSFUL

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'
charity_data_df.drop(['EIN', 'NAME'], axis = 1, inplace = True)
charity_data_df.head(5)
```

```python
# Determine the number of unique values in each column.
charity_data_df.nunique()
```

```python
# APPLICATION_TYPE value counts for binning
app_count = charity_data_df['APPLICATION_TYPE'].value_counts()
app_count
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# Use the variable name `application_types_to_replace`
application_types_to_replace = list(app_count[app_count < 156].index)

# Replace app types with less than 156 apps as Other in dataframe
for app in application_types_to_replace:
    charity_data_df['APPLICATION_TYPE'] = charity_data_df['APPLICATION_TYPE'].replace(app,"Other")

# Verify binning.
charity_data_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = charity_data_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# Use the variable name `classifications_to_replace`
classifications_to_replace = list(classification_counts[classification_counts < 1000].index)

# Replace with Other in dataframe
for cls in classifications_to_replace:
    charity_data_df['CLASSIFICATION'] = charity_data_df['CLASSIFICATION'].replace(cls,"Other")
    
# Verify binning
charity_data_df['CLASSIFICATION'].value_counts()
```

```python
numeric_charity_df = pd.get_dummies(charity_data_df)
numeric_charity_df.head(5)
```

```python
# Split our preprocessed data into our features and target arrays
y = numeric_charity_df['IS_SUCCESSFUL']
X = numeric_charity_df.drop(['IS_SUCCESSFUL'], axis=1)

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### **Compile, Train, and Evaluate the Model**


```python
# Create a neural network model by assigning the number of input features and nodes for each layer using Tensorflow Keras
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
nn = tf.keras.models.Sequential()
nn.add(tf.keras.layers.Dense(units=10, activation="relu", input_dim= number_input_features, name = 'layer1')) 
nn.add(tf.keras.layers.Dense(units=15, activation="relu", name = 'layer2'))
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid")) 
nn.summary()

```

```python
# Compile the model
# Using adam optimizer to minimize loss
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
# Create a callback that saves the model's weights every 5 epochs.
callback = tf.keras.callbacks.ModelCheckpoint(filepath = "Models/weights.{epoch:05d}.hdf5", save_freg = "epoch", period=5)
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Attempt to optmize by adding another layer to reach accuracy of 75%

from keras.models import Model

layer_name = 'dense_1'
model2= Model(inputs=nn.input, outputs=nn.get_layer(layer_name).output)
model2.summary()
```

```python
# Compile the model
# Using adam optimizer to minimize loss
model2.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
# Create a callback that saves the model's weights every 5 epochs.
callback = tf.keras.callbacks.ModelCheckpoint(filepath = "Models/weights.{epoch:05d}.hdf5", save_freg = "epoch", period=5)
```

```python
fit_model = model2.fit(X_train_scaled, y_train, epochs=100)
```

```python
model_loss, model_accuracy = model2.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
